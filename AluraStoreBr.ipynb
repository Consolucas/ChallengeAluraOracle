### Importação dos dados



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja2.head()
```

#1. Análise do faturamento


```python
"""teste"""
loja1['n vendas'] = loja1['Vendedor'].count()
loja1
```

```python
"""teste"""
loja1 = loja1.drop('n vendas', axis=1)
loja1
```

```python
lojas = [loja1, loja2, loja3, loja4]
faturamento = {}
for i in range(len(lojas)):
  faturado = lojas[i]['Preço'].sum()
  faturamento[f'loja{i+1}'] = f"R${faturado:_.2f}".replace('.', ',').replace('_', '.')
faturamento
```

```python
lojas_graf = []
faturamento_graf = []
for i in range(len(lojas)):
  faturamento_graf.append(float(lojas[i]['Preço'].sum()))
  lojas_graf.append(f'loja{i+1}')
plt.bar(lojas_graf, faturamento_graf)
plt.title('Faturamento por loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
lojas_graf = []
faturamento_graf = []
for i in range(len(lojas)):
  faturamento_graf.append(float(lojas[i]['Preço'].sum()))
  lojas_graf.append(f'loja{i+1}')
plt.bar(lojas_graf, faturamento_graf)
for i, valor in enumerate(faturamento_graf):
    plt.text(i, valor, str(valor), ha='center', va='bottom')
plt.title('Faturamento por loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

# 2. Vendas por Categoria


```python
'''teste'''
categorias = list(loja1['Categoria do Produto'].sort_values().unique())
result = []
for i in range(len(lojas)):
  for x in range(len(categorias)):
    list_temp = []
    list_temp.append(f'loja{i+1}')
    list_temp.append(categorias[x])
    count = (float(lojas[i].loc[lojas[i]['Categoria do Produto'] == categorias[x]]['Preço'].sum()))
    list_temp.append("{:.2f}".format(count))
    result.append(list_temp)
result
```

```python
'''teste'''
categorias = list(loja1['Categoria do Produto'].sort_values().unique())
result = {}
for i in range(len(lojas)):
  for x in range(len(categorias)):
    list_temp = []
    loja = f'loja{i+1}'
    list_temp.append(loja)
    list_temp.append(categorias[x])
    count = (float(lojas[i].loc[lojas[i]['Categoria do Produto'] == categorias[x]]['Preço'].sum()))
    list_temp.append("{:.2f}".format(count))
    if loja not in result:
      del list_temp[0]
      result[loja] = [list_temp]
    else:
      del list_temp[0]
      result[loja].append(list_temp)
result
```

```python
categorias = list(loja1['Categoria do Produto'].sort_values().unique())
result = {}
for i in range(len(lojas)):
  result[f'loja{i+1}'] = {}
  for x in range(len(categorias)):
    count = (float(lojas[i].loc[lojas[i]['Categoria do Produto'] == categorias[x]]['Preço'].sum()))
    result[f'loja{i+1}'][categorias[x]] = "{:.2f}".format(count)
    print(result)
```

```python
for i in result:
  rotulos = result[i].keys()
  valores = result[i].values()
  plt.pie(valores, labels=rotulos, autopct='%1.1f%%', startangle=140)
  # Adicionar título
  plt.title('Distribuição de Dados')

  # Ajustar a proporção para que o gráfico seja um círculo
  plt.axis('equal')

  # Exibir o gráfico
  plt.show()
```

# 3. Média de Avaliação das Lojas

```python
for i in range(len(lojas)):
  media = lojas[i]['Avaliação da compra'].mean()
  media_formatada = "{:.2f}".format(media)
  print(f'Loja{i+1}: {media_formatada}')
```

# 4. Produtos Mais e Menos Vendidos

```python
'''teste'''
c = 0
for i in lojas:
  for x in i['Produto']:
    if x == 'Iniciando em programação':
      c += 1
  print(c)
```

```python
for i in range(len(lojas)):
  mais_vendido = lojas[i].groupby('Produto')['Preço'].count().idxmax()
  menos_vendido = lojas[i].groupby('Produto')['Preço'].count().idxmin()
  print(f'Loja{i+1}:')
  print(f'Mais vendido: {mais_vendido}')
  print(f'Menos vendido: {menos_vendido}')
```
